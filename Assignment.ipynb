Importing Required Library and reading csv file

Creating a list to store working days and time.

```python
l = [a for a in df["Date"].unique()]
l1 = [a for a in df["Time"].unique()]
```

```python
vol = []
```

```python
rank=[]
```

Now for every day on each minute, the volume is being compared to the last 5 days to find out the rank.

```python
for i in l:
    for k in l1:
        volume = df[(df["Date"] == i) & (df['Time'] == k)]['Volume'].values[0]
        vol=[]
        for j in range(l.index(i), l.index(i)-6, -1):
            if j < 0:
                break
            vol.extend(df[(df["Date"] == l[j]) & (df['Time'] == k)]['Volume'].values)
        vol.sort(reverse=True)
        a = vol.index(volume)+1
        rank.append(a)
```

```python
len(rank)
```

```python
df['rank'] = rank
```

```python
pd.options.display.max_rows = 10000
```

Rank is now added to the original DataFrame

Plotting a graph of date 19-01-2024 for every minute where rank is 1 

```python
df1 = df[(df['Date']=='19-01-2024') & (df['rank'] == 1)]
```

```python
plt.figure(figsize=(14,8),dpi=200)
plt.plot(df1['Time'], df1["Volume"])
plt.xticks(rotation=90);
```

On 19-01-2024 Peak Volume from last 5 days (rank 1) is at 11:06:00 and Lowest volume at 14:11:00 but still highest volume from last 5 days.
